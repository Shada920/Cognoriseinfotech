```python
selected_cols=df[['price','bedrooms' , 'bathrooms','sqft_living' , 'sqft_lot','floors' , 'waterfront' , 'view' , 'condition' , 'sqft_above' , 'sqft_basement' , 'yr_built','yr_renovated']]
corr_matrix=selected_cols.corr(method="pearson")
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
high_corr_feature=corr_matrix.index[abs(corr_matrix['price'])>=0.2].tolist()
high_corr_feature.remove('price')
print(high_corr_feature)
```

```python
y=df['price']
x=df[high_corr_feature]
scaler = StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_test_pre=reg.predict(x_test)
test_acc=r2_score(y_test,y_test_pre)
test_acc
```

```python
cutoff = int(len(df) * 0.8) # 4600 * 0.8

# getting first 80% data for training
df_train = df[:cutoff]

# getting last 20% data for testing
df_test = df[cutoff:]
```

```python
X_train = df_train[['bedrooms', 'bathrooms', 'sqft_living', 'floors']]
X_test = df_test[['bedrooms', 'bathrooms', 'sqft_living', 'floors']]

# our output
y_train = df_train['price']
y_test = df_test['price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
equation = ""
for i, column in enumerate(X_train.columns):
    equation =   f"({m[i]} x {column})" + " + " + equation

equation = equation + f"{c}"

print("Our Equation is : \n", equation)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"True price is: {list(y_test)[1]} a predicted price is {y_pred[1]}")
print(f"True price is: {list(y_test)[2]} a predicted price is {y_pred[2]}")
print(f"True price is: {list(y_test)[3]} a predicted price is {y_pred[3]}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

baseline_mae = mean_absolute_error([y_train.mean()] * len(y_train), y_train)

pred_mae = mean_absolute_error(y_test, y_pred)
pred_mse = mean_squared_error(y_test, y_pred)
pred_mae = mean_absolute_error(y_test, y_pred)

print(f'baseline {baseline_mae}')
print(f'baseline {pred_mae}')
print(f'baseline {pred_mse}')
```
